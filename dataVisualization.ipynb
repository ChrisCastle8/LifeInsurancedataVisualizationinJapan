```python
import numpy as np # linear algebra
import pandas as pd # data processing, csvfile in/output
# vizualisation tools
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import re
import calendar
import scipy.stats as stats 
import warnings
df2020 = pd.read_csv("2020lifeInsurance.csv")
df2019 = pd.read_csv("2019lifeInsurance.csv")
df2018 = pd.read_csv("2018lifeInsurance.csv")
df2017 = pd.read_csv("2017lifeInsurance.csv")
```

```python
plt.figure(figsize = (12,6))
sns.lineplot(x = df2020.Month , y = df2020.PremiumIncome, color = "green", label = "2020",linewidth=4)
sns.lineplot(x = df2019.Month , y = df2019.PremiumIncome, color = "crimson", label = "2019",linewidth=4)
sns.lineplot(x = df2018.Month , y = df2018.PremiumIncome, color = "navy", label = "2018",linewidth=4)
sns.lineplot(x = df2017.Month , y = df2017.PremiumIncome, color = "orange", label = "2017",linewidth=4)
plt.xlabel('Months of the Year', fontsize = 15)
plt.ylabel('Premium Income in Billions of Yen', fontsize = 15)
plt.title('Premium Income Comparison from 2017 to 2020', fontsize = 20)

plt.savefig('premiumincomecomparison')
```

```python
corr = df2020.corr()
plt.figure(figsize=(12,7))
sns.heatmap(corr, annot = False, cmap = 'gist_stern_r')
plt.title('Correlation between Revenues and Expenditures in 2020', fontsize = 20)
plt.savefig('final 2020 heatMap')
```

```python
plt.figure(figsize = (12,8))
plt.rcParams['font.size'] = '14'
sns.barplot(x = df2020.Month , y = df2020.PremiumIncome, color = "green")
plt.xlabel('Months of the Year', fontsize = 15)
plt.ylabel('Premium Income in Billions of Yen', fontsize = 15)
plt.title('Monthly Premium Income Comparison for 2020', fontsize = 20)
plt.savefig('Final 2020monthlypremiumincomecomparison')
```
